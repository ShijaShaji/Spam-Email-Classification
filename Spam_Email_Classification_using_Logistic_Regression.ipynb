```python
df["class"].replace({0:-1}, inplace=True)
```

```python
#Split the data to train and test in the ratio 80-20
trainData = df.sample(frac=0.8, random_state=25)
testData = df.drop(trainData.index)
```

```python
npTraindata = trainData.to_numpy()
np.random.seed(5)                                      #permuting the dataset in each iteration for better results
npTraindata = np.random.permutation(npTraindata)
npTestData = testData.to_numpy()
```

```python
X_Train = npTraindata[:,:57]
Y_Train = npTraindata[:,57]
X_Test = npTestData[:,:57]
Y_Test = npTestData[:,57]
```

```python
Y_Train
```

```python
def GaussianNormalisation(dataset):
    #Compute the number of features
    numFeatures = len(dataset[0])
    
    featureMean = np.empty(numFeatures, float)
    featureStd = np.empty(numFeatures, float)
    
    #For every feature
    for i in range(numFeatures):
        #find its Mean and Std
        featureMean[i] = dataset[:,i].mean(axis=0)
        featureStd[i] = dataset[:,i].std(axis=0)
        #Apply Gaussian Noramlisation
        dataset[:,i] = (dataset[:,i] - featureMean[i])/featureStd[i]

    return (featureMean, featureStd)
```

```python
(featureMean, featureStd) = GaussianNormalisation(X_Train)
```

```python
# Implement sigmoid function
def sigmoid(x):
    # Bound the argument to be in the interval [-500, 500] to prevent overflow
    x = np.clip( x, -500, 500 )

    return 1/(1 + np.exp(-x))
```

```python
def logisticRegression(features,labels):
  MaxIter = 20
  lr = 0.01
  b = 0
  numObjects,numFeatures =np.shape(features)
  W = np.zeros(numFeatures)   

  for i in range(1,MaxIter+1):                                                
      for j in range(numObjects):  
         X = features[j] 
         y = labels[j]                   
         a = (np.dot(W.T,X)) +  b                   #calculating activation score
         #update bias and weights   
         W = np.add(W,(lr*y*sigmoid(-y*a)*X))
         b = b + (lr*y*sigmoid(-y*a))

  return b,W
```

```python
bias , weights = logisticRegression(X_Train, Y_Train)
print(bias,weights)
```

```python
##normalize the test dataset using Means and Std computed on the training dataset
#Compute the number of features
numFeatures = len(X_Test[0]) 
   #For every feature
for i in range(numFeatures):
      #Apply Gaussian Noramlisation with given Mean and Std values
    X_Test[:,i] = (X_Test[:,i] - featureMean[i])/featureStd[i]
```

Predict the classes of test dataset using the computed bias and weights

```python
def logisticRegressionTest(b, W, obj):
    #Compute the activation score
    a = np.dot(obj, W) + b
    predictedClass = 0
    confidence = 0
    if a > 0:
        predictedClass = +1
        confidence = sigmoid(a)
    else:
        predictedClass = -1
        confidence = 1-sigmoid(a)
    return (predictedClass, confidence)
```

```python
def evaluationReport(classTrue, classPred):
    positive_mask = classTrue == 1

    # Count the number of elements in the positive class 
    positive = np.count_nonzero(positive_mask)
    # Count True Positive
    tp = np.count_nonzero(classPred[positive_mask]==1)
    # Count False Negative
    fn = np.count_nonzero(classPred[positive_mask]==-1)
    
    negative_mask = classTrue == -1

    # Count the number of elements in the negative class 
    negative = np.count_nonzero(negative_mask)
    # Count False Positive
    fp = np.count_nonzero(classPred[negative_mask]==1)
    # Count True Negative
    tn = np.count_nonzero(classPred[negative_mask]==-1)

    # Compute Accuracy, Precision, Recall, and F-score
    accuracy = (tp + tn)/(tp + tn + fp + fn)
    precision = tp/(tp + fp)
    recall = tp/(tp + fn)
    fscore = 2*precision*recall/(precision + recall)
    print("Evaluation report")
    print("Accuracy: %.2f" % accuracy)
    print("Precision: %.2f" % precision)
    print("Recall: %.2f" % recall)
    print("F-score: %.2f" % fscore)
```

```python
classTrue = np.array([int(x) for x in Y_Test], dtype=int)
classPred = np.array([int(logisticRegressionTest(bias,weights,X)[0]) for X in X_Test], dtype=int)
evaluationReport(classTrue, classPred)
```
